 ### **Tuples packing and unpacking**

Tuple packing is when multiple values are assigned to a single variable, and Python automatically groups these values into a tuple. For example, data_pack = 1, 2, 'three' creates a tuple (1, 2, 'three'). Tuple unpacking is the reverse: it allows you to assign the elements of a tuple to multiple variables in a single statement, like a, b, c = data_pack. The number of variables on the left side of the assignment must match the number of elements in the tuple.

```python
#tuples packing in python
data_pack = 1,2, "three"
data_pack
print(f"data_pack = {data_pack}")
```

```python
#tuples unpacking in python
a,b,c = data_pack
print(f"a = {a}")
print(f"b = {b}")
print(f"c = {c}")
```

```python
#wants only two value
a,b,_ = data_pack
print(f"a = {a}")
print(f"b = {b}")
```

```python
#skippin the second value
a,_,c = data_pack
print(f"a = {a}")
print(f"c = {c}")
```
