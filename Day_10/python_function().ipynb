```python
# class task 2
def Calculate(a,b,c):
  if (c == "+"):
    return a+b
  elif (c == "-"):
    return a-b
  elif (c == "*"):
    return a*b
  elif (c == "/"):
    return a/b
  else:
    return "Invalid sign"

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
sign = input("Enter sign: ").strip()

d = Calculate(a,b,sign)
print("After performing the calculation: ",d)
```

```python
#class task 3
def udf():
  l = []
  for i in range(5):
    n = input("Enter a number: ")
    if "." in n:
      if (len(l)>0):
        remove = l.pop()
        print(f"float detected. Removed: {remove}")
      else:
        print("float detected but the list is empty")
    else:
      l.append(int(n))
      print(f"{n} added to the list")
  return l

c = udf()
print(c)
```
