```python
#Creating dictionary
student ={
    "name": "Tony",
    "age" : 22,
    "course" : "dsml"
}

print(f"studet data = {student}")
print(f"student name = {student['name']}")
print(f"student age = {student['age']}")
print(f"student course = {student['course']}")
```

```python
#creating dictionary using dict()
dict_data =  dict({
    "name" : "richard",
    "home" : "usa",
    "phone" : 65445554,
    "courses" : ["dsml", "python", "java"] #list inside dictionary
})
print(f"dict_data = {dict_data}")
print(f"dict_data name = {dict_data['name']}")
print(f"dict_data home = {dict_data['home']}")
print(f"dict_data phone = {dict_data['phone']}")
print(f"dict_data courses = {dict_data['courses']}")
```

```python
#dictionary concationation
dict1 ={
    "name":"stard",
    "age" :22
}
dict2 ={
    "name":"tony", # If provides same key than later value provided will overwrite
    "city":"btm",
    "state":"karnataka"
}
dict1.update(dict2)
print(f"dict1 = {dict1}")
print(f"dict2 = {dict2}")
#print(f"dict1 + dict2 = {dict1 + dict2}") # won't work
print(f"dict1.update(dict2) = {dict1}")
```

```python
#using #kwargs to unpack
dict3 ={
    "gender" : "male",
    "marks" : 45
}

st_data = {**dict1,**dict2, **dict3}
print(f"st_data = {st_data}")
#
```

```python
#sort
result = sorted(st_data)
print(f"result = {result}")
```

```python
#merge dictionary
dict1 | dict2 | dict3
```
