Shallow copy vs deep copy

```python
lst = [[1,2,3,4],[23,3,4,6,7]]
lst1 = lst
```

```python
lst2 = lst.copy()
```

```python
lst[1][1] =59 #soft copy
print(lst)
print(lst2)
```

```python
#deep copy
import copy
```

```python
datas = [[1,2,3,4],[23,3,4,6,7]]
datas1 = copy.deepcopy(datas)
```

```python
datas[1][1] =100
```

```python
print(datas)
print(datas1)
```
