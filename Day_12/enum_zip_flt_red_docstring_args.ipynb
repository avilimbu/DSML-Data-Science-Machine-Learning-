```python
#zip
data1 = ["ram", "shyam", "gita", "rita"]
data2 = [1,2,3,4]
for i,j in zip (data1,data2):
  print(i)
```

```python
for i, (name,age) in enumerate(zip(data1,data2)):
  print(i,name,age)
```

```python
#class task
names = ["ram", "shyam", "gita", "rita"]
age = [22,23,34,43]
gender= ["male","male","female","female"]
salaries = [10000, 20000, 30000, 4500]
for i, (name,age,gender,salary) in enumerate(zip(names,age,gender,salaries), start=1):
  print(f"Name = {name},Age = {age}, Gender = {gender}, Salary = {salary}")
```

```python
from typing import Iterable
#map (function, iterable)
def mux(x):
  return x*x
number = [4,5,7,9]
map_obj = map(mux,number)
print(list(map_obj))
```

```python
# filter
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
def even(x):
  if (x%2==0):
    return x

even_numbers = list(filter(even, numbers))

print(even_numbers)
```

```python
#reduce
from functools import reduce
my_lst = [1,2,3,4,5]
def mn(x,y):
  return x*y
print(reduce(mn,my_lst))
```

```python
my_lists = [45,43,88,65]
def max(x,y):
  return x if x>y else y
print(reduce(max,my_lists))
```

```python
#positional argument
def calculate(*args):
    if len(args) == 2:
        l, b = args
        print("Area =", l * b)

    elif len(args) == 3:
        l, b, h = args
        print("Volume =", l * b * h)

    else:
        print("Please provide either 2 or 3 arguments.")

calculate(3, 4)
calculate(3, 4, 5)
```
