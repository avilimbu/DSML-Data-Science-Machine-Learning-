```python
#recursin
def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)
factorial(8)
```

```python
def fibonacci(n):
    if n<=1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input("Enter how many Fibonacci numbers to print: "))
if (n<=0):
  print("enter a valid number of series.")
else:
  for i in range(n):
    print(fibonacci(i), end=",")
```

```python
#Nested function
def outer(x):
  def inner(y):
    return x+y
  return inner
add_= outer(5)
print(add_)
```

```python
#return function as a value
def outer(name):

    def inner():
        print("Hello,", name)

    return inner      # Return the function, don't call it

message = outer("Avik")
print(message)
message()
```

```python
#decoretor
def myWrapper(func):
  def myInnerFunc():
    print("Inner Wrapper.")
    func()
  return myInnerFunc

@myWrapper
def myFunc():
  print("hello there!")

myFunc()
```

```python
def logger(func):
  def wrapper(*args,**kwargs):
    print(f"Calling {func.__name__}")
    return func(*args,**kwargs)
  return wrapper
@logger
def greet(name,age):
  print(f"Hello {name}, Your age: {age}")

greet("avhi", age=22)
```
